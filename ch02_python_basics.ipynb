## Control Flow

### if, elif, and else

```python
x = -5
if x < 0:
    print("It's negative")
```

```python
if x < 0:
    print("It's negative")
elif x == 0:
    print("Equal to zero")
elif 0 < x < 5:
    print("Positive but smaller than 5")
else:
    print("Positive and larger than or equal to 5")
```

```python
a = 5; b = 7
c = 8; d = 4

if a < b or c > d:
    print("Made it!")
```

```python
# Chain comparisons
4 > 3 > 2 > 1
```

### for loops

```python
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
    if value is None:
        continue
    total += value
```

```python
total
```

```python
sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0
for value in sequence:
    if value == 5:
        break
    total_until_5 += value
```

```python
total_until_5
```

### while loops

```python
x = 256
total = 0
while x > 0:
    if total > 500:
        break
    total += x
    x = x // 2
```

```python
total
```
